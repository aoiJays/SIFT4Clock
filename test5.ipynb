```python
import cv2
import matplotlib.pyplot as plt

# 
image = cv2.imread('source/1.2/tv20171227162553V75UJF.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 
median_filtered = cv2.medianBlur(gray, 5)
# 
adaptive_thresh = cv2.adaptiveThreshold(median_filtered, 255, 
                                        cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                        cv2.THRESH_BINARY, 
                                        11, 2)

# 
plt.figure(figsize=(32, 32))
plt.imshow(adaptive_thresh, cmap='gray'), plt.title('Edge Detected')
plt.show()
```
